```python
import numpy as np
import matplotlib.pyplot as plt

# Function to simulate predator-prey dynamics using the Lotka-Volterra model
def simulate_ecosystem(preys, predators, growth_rate_preys, growth_rate_predators, interaction_coeff_a, interaction_coeff_b, dt, steps):
    prey_population = [preys]
    predator_population = [predators]

    for _ in range(steps):
        dH = (growth_rate_preys * prey_population[-1] - interaction_coeff_a * predator_population[-1] * prey_population[-1]) * dt
        dP = (-growth_rate_predators * predator_population[-1] + interaction_coeff_b * predator_population[-1] * prey_population[-1]) * dt

        preys += dH
        predators += dP

        prey_population.append(preys)
        predator_population.append(predators)

    return prey_population, predator_population

# Initial conditions and parameters
initial_preys = 40
initial_predators = 9
growth_rate_preys = 0.1
growth_rate_predators = 0.02
interaction_coeff_a = 0.01
interaction_coeff_b = 0.005
time_step = 0.1
simulation_steps = 500

# Run the simulation
prey_population, predator_population = simulate_ecosystem(
    initial_preys, initial_predators,
    growth_rate_preys, growth_rate_predators,
    interaction_coeff_a, interaction_coeff_b,
    time_step, simulation_steps
)

# Fixing the dimensions to make time_points and populations have the same length
time_points = np.arange(0, (simulation_steps + 1) * time_step, time_step)
prey_population = np.array(prey_population)[:len(time_points)]
predator_population = np.array(predator_population)[:len(time_points)]

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(time_points, prey_population, label='Prey (H)')
plt.plot(time_points, predator_population, label='Predator (P)')
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('Predator-Prey Simulation')
plt.legend()
plt.grid(True)
plt.show()

```
